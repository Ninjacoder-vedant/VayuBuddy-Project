#### Use below code to create dataset from your question_templates.json

```python
import pandas as pd
import itertools
```

```python
# Load the JSON file into a pandas DataFrame.
question_templates = pd.read_json("question_templates.json")

# This list will store all the final generated question items.
question_list = []

# Loop through every row (template) in the DataFrame.
for i in range(len(question_templates)):
    # Get the current row as a Series.
    row = question_templates.iloc[i]
    
    # Extract the question template and the code template from the row.
    ques_template = row['question_template']
    code_template = row['code_template']
    
    # Get the list of variable names for the question and the code from the row.
    ques_vars = row['ques_vars']
    code_vars = row['code_vars']
    
    # Also extract the category of the template.
    category = row['category']
    
    # For the question template, build a list of lists, where each sublist contains 
    # all possible values for the corresponding variable.
    ques_vals = []
    for var in ques_vars:
        ques_vals.append(row[var])
    
    # For the code template, do the same: build a list of lists of possible values.
    code_vals = []
    for var in code_vars:
        code_vals.append(row[var])
    
    # Generate the Cartesian product (i.e., all combinations) of question values.
    ques_combinations = itertools.product(*ques_vals)
    
    # Generate the Cartesian product of code values.
    code_combinations = itertools.product(*code_vals)

    ques_combos_list = list(ques_combinations)
    code_combos_list = list(code_combinations)
    
    id = 0
    # Here, zip() pairs the combinations one-to-one.
    for q_combo, c_combo in zip(ques_combos_list, code_combos_list):
            id += 1 
            # Create a dictionary mapping question variable names to their chosen values.
            ques_fill = dict(zip(ques_vars, q_combo))
            # Create a dictionary mapping code variable names to their chosen values.
            code_fill = dict(zip(code_vars, c_combo))
            # Replace the placeholders in the question template with the actual values.
            filled_ques = ques_template.format(**ques_fill)
            # Replace the placeholders in the code template with the actual values.
            filled_code = code_template.format(**code_fill)
            
            # Append the generated question and its details to the question_list.
            question_list.append({
                "category": category,
                "question": filled_ques,
                "true_code": filled_code
            })
questions_df = pd.DataFrame(question_list)
# save the questions to csv file
questions_df.to_csv("questions.csv", index=False)
```

#### Running the codes to check if the template is giving correct codes

```python
import subprocess
import concurrent.futures  # This module makes it easy to launch parallel tasks.

# Convert the 'true_code' column into a list of code snippets.
codes = questions_df['true_code'].tolist()

def execute_code(code):
    try:
        # Execute the code using subprocess.check_output
        # The stderr=subprocess.STDOUT argument combines the stdout and stderr so that if there is an error, 
        # you get the error output.
        output = subprocess.check_output(["python3", "-c", code], stderr=subprocess.STDOUT)
        # Decode from bytes to string and remove any whitespace.
        return output.decode("utf-8").strip()
    except subprocess.CalledProcessError as e:
        # If the executed code fails (non-zero exit), return the error message.
        return "Error: " + e.output.decode("utf-8").strip()

# Create a ProcessPoolExecutor to run the execute_code function in parallel.
with concurrent.futures.ProcessPoolExecutor() as executor:
    # Map the list of code snippets to the execute_code function.
    # This will distribute code execution across available CPU cores.
    results = list(executor.map(execute_code, codes))

# Store the results back into the DataFrame as a new column 'correct_ans'.
questions_df['correct_ans'] = results
```
